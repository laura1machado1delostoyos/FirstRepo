Assume an assignment where you want to go up the Empire State Building (100 floors) by rolling a die.
The rules are:
* If the die outcome is 1 or 2, you go down one floor
  * Assuming you are not in floor zero -you can't go udnerground!
* If the die outcome is 3, 4 or 5, you go up one floor
* If the die outcome is six, you throuw it again and then walk up however many floors you got as a result.
* You have a 0.1% chance of falling everytime you roll the die.

```python
import random
```

Since the chance of falling is 0.1%, I wanted to create a list to draw from where 1 out of 1000 items was the string `'Fall!'`. In the if/else statements, this would trigger the `current_floor` variable to be updated to zero.
This would then mean that the remaining 999 items of the list would be randomly populated by the items 1 through 6.
However, 999 is not divisible by 6. The closest number to 999 that is divisible by 6 is 996. So this would mean that whatever numbers occupied the 997th, 998th and 999th position would have a slightly greater incidence than the rest, which results in a biased die.
It turns out, however, that 1998, which is 999 x 2, is in fact divisible by 6.
So, instead of making a list of 1,000 items, I will built a list of 2,000 items where:
* 1,998 items are the numbers 1 through 6 (with even weights)
* 2 items are the string `'Fall!'`

```python
# Lets code all possible 6 outcomes from a die
die_outcomes = [i + 1 for i in range(6)]
print(die_outcomes)
print(len(die_outcomes))
```

```python
# Lets extend that list to 1,998 positions
die_outcomes_1998 = []

while len(die_outcomes_1998) < 1998:
  die_outcomes_1998.extend(die_outcomes)

print(die_outcomes_1998)
print(len(die_outcomes_1998))
```

```python
die_outcomes_plus_falls = die_outcomes_1998.copy()
die_outcomes_plus_falls.append('Fall!')
die_outcomes_plus_falls.append('Fall!')
print(die_outcomes_plus_falls)
print(len(die_outcomes_plus_falls))
```

```python
max_throws = 100
max_floors = 100
current_floor = 0
current_throws = 0


# We only want to throw the dice when we are not on the last floor, and within our assigned maximum tries
while current_throws < max_throws and current_floor < max_floors:
  # Lets print a statement before each throw so that the reader can position us.
  print('We have thrown the dice ' + str(current_throws) + ' times, and we are in floor no. ' + str(current_floor))
  print()

  # Lets throw the dice!
  dice_outcome = random.choice(die_outcomes_plus_falls)

  # And print the outcome of the dice being thrown
  print('On throw no. ' + str(current_throws + 1) + ', the dice outcome was ' + str(dice_outcome)) # see the result
  current_throws += 1 #counting no. throws

  # if/else statements for each of the possiible outcomes and their consequences

  if dice_outcome in [1,2]:
    current_floor = max(0, current_floor - 1)
    print('Womp womp! We got a ' + str(dice_outcome) + ' so it means we go down to floor no. ' + str(current_floor))

  elif dice_outcome in [3, 4, 5]:
      current_floor += 1
      print('Nice! We got a ' + str(dice_outcome) + ' so it means we go up to floor no. ' + str(current_floor))

  elif dice_outcome == 6:
    current_throws += 1 #counting no. throws
    dice_outcome = random.choice(die_outcomes_plus_falls)

    if dice_outcome in list(range(7)): #dice_outcome will never be zero, but this is in the interest of time

      current_floor = min(max_floors, current_floor + dice_outcome)

      print('We got a six! We played again and got a ' + str(dice_outcome) + ' so the new floor number is ' + str(current_floor)) # see the result

    else:
      current_floor = 0
      print('We got a six but after playing again we fell.')

  else:
    print('Nooooo! We fell.')
    current_floor = 0
# Once we exit the loop, lets see a print statement to see whether it was due to us going over our maximum no. of throws
# or whether it was due to arriving at the top
print()
print('---Summary---')
print('We threw the dice ' + str(current_throws) + ' times, and ended up on floor ' + str(current_floor))
```
